```python
from operator import index
from pandas.io.json import json_normalize
import json as j
import pandas as pd
import requests as r 
import itertools

#prendo la prima parte del url per la api
base_url = "http://ergast.com/api/f1/circuits.json"

URL0= r.get(base_url)

info=URL0.json()

#Ottengo il numero massimo di elementi dei circuiti
def get_key(key):
    return [entry[key] for entry in info.values()]
gk=(get_key('total'))
total=gk[0]

#URL della API Completa
URL1="http://ergast.com/api/f1/circuits.json?limit="+total

#effettuo la connessione aggiornata
URL=r.get(URL1)

#import i dati in json totali
circuits_info=URL.json()


#Ottengo le info dei circuiti
def get_circuits(constrkey):
   return [entry[constrkey] for entry in circuits_info.values()]
dk1=(get_circuits('CircuitTable'))
circuits_info_total1=dk1
df=pd.DataFrame(circuits_info_total1)

def get_constr2(constrkey2):
    return [entry[constrkey2] for entry in circuits_info_total1]
dk2=(get_constr2('Circuits'))
circuits_info_total2=dk2

#Cambio da lista a variabile da dare a pandas
circuits_info_total=list(itertools.chain(*circuits_info_total2))

#informazioni in dataframe
df1=pd.DataFrame(circuits_info_total)
df1
# salvo le informazioni già utili
circuits1_2=df1[["circuitId","circuitName"]]


#La lista dentro gli assegno il nome df2
df2=df1["Location"]

#Normalizzo la lista
circuits2_2=pd.json_normalize(df2)

#Concateno e rinomino le lat e le long in 'lat,long'
circuits2_3=pd.DataFrame({"lat,long" : circuits2_2["lat"].astype(str)+","+circuits2_2["long"].astype(str)})

#Aggrego le colonne
circuits=pd.concat([circuits1_2,circuits2_2["locality"],circuits2_2["country"],circuits2_3], axis=1)
circuits
```
